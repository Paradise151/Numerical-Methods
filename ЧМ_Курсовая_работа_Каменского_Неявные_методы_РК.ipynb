# Теоретическая часть, анализ проблематики использования явных методов в решенении жёстких систем ОДУ

# *Определение*

В связи с наличием у некоторых решений систем ОДУ компонент, имеющих резкий рост в окрестности некоторых точек (пограничного слоя), вводится понятие числа жёсткости для систем ОДУ с $\lambda_i < 0$  \\
$$ s = \frac{max|\lambda_{i}|}{min|\lambda_{i}|}$$

Соотвественно система ОДУ называется жёсткой, если у неё большое число жёсткости.

# *Анализ проблематики*

Известно, что метод Эйлера численного решения ОДУ имеет низкую точность - в основном он служит для иллюстрации работы численных методов решения диффуров. Но всё же рассмотрим применение вышеописанного метода для следующего НЕОДНОРОДНОГО ОДУ. \\
$$y'=-\lambda\,\left(y-\sin\left(x\right)\right)$$  

image.png

Как видим, полученное решение очень плохо аппроксимирует аналитическое решение.
 Связано это с началичием резкого убывания решения вблизи х=0 (в таком случае x = 0 называют пограничным слоем). Резкое убыванием обусловлено тем, что при больших $k$ и малых $t$ имеем  $y\approx e^{-\lambda\,x}$. Само решение ОДУ имеет следущий вид:  \\
$$y=\dfrac{{\lambda}^{2}\,\sin\left(x\right)}{{\lambda}^{2}+1}-\dfrac{\lambda\,\cos\left(x\right)}{{\lambda}^{2}+1}+\dfrac{{\lambda}^{2}+\lambda+1}{({\lambda}^{2}+1)e^{\lambda\,x}}$$

Теперь рассмотрим то же ОДУ, но задача Коши будет поставлена для точки, находящейся вне интервала резкого убывания. Выберем, например, точку (0.5, 0.435)

# Неявные методы решения жёстких систем ОДУ

# Постановка задачи для численного решения неявными методами РК

Неявные методы хорошо справляются с жёсткими системами, так как в отличии от явных избегают использование исключительно данных предыдущей точки (например явный метод Эйлера испольузует тот факт, что при малом шаге касательная не успевает слишком сильно разойтись с функцией аналитического решения).

Рассмотрим уравнение колебаний осциллятора с демпфированием. \\
$y''=-\left(k+1\right)\,y'-k\,y; \quad y(0)=1, y'(0)=0; \quad (c=k+1, k>0) \\
  \begin{cases}
  y'=z   \\
  z'=y''= f(x,y,z) = -(k+1)z-ky
   \end{cases}
   $

Если анализировать жёсткость визуально по графику решению, то у функции Y интервала резкого убывания нет, но погрешность численного решения всё же оказывается слишком большой. Это связано с наличием резкого убывания у производной функции Y. Что в свою очредь связано с жесткостью системы равной $$ s = \frac{max|\lambda_{i}|}{min|\lambda_{i}|}=k$$ Этим хочется подчернкуть важность определения жёсткости через собственные числа матрицы системы ОДУ.

Решение задачи:

$y''=-\left(k+1\right)\,y'-k\,y\; \\
  \begin{cases}
  y=-\dfrac{1}{{(k-1)e}^{k\,t}}+\dfrac{k}{{(k-1)e}^{t}}   \\
  z=y'=-\dfrac{k}{{(k-1)e}^{-k\,t}}-\dfrac{k}{{(k-1)e}^{-t}}
   \end{cases}
   $

Как видим пограничный слой имеется у производной функции Y. Связано это с тем, что при больших $k$ и малых $t$ имеем $z\approx e^{-kt}-1$ и $y\approx e^{-t}$. Поэтому в полуокрестности 0 - абциссы поля, имеется сильное убывание, которое влияет на явные методы (значения касательной и функции в точке $x_k + \Delta h$ слишком сильно отличаются).

# Неявный метод Эйлера

Хотелось, конечно, использовать метод Ньютона, ведь имеется система нелинейных уравнений, но для каждого прохода цикла в методе Эйлера необходимо приближенное решение. Но в ущерб быстроте (потребуется большее число итарций)  можно для всех проходов цикла (т.е для каждой системы) в качестве начального приближения выбрать точку, соответующую серидине графика функции Y.

$\begin{cases}
  y_{k+1}= y_k + h \cdot z_{k+1}  \\
  z_{k+1}= z_k + h \cdot f(x_{k+1},y_{k+1},z_{k+1})
   \end{cases}
   $

Неявный метод Эйлера является устойчивым для данного типа задач с отрицательными собственными значениями (А-устойчивый).

```python
def zeros2(n,m):
  return [m*[0] for _ in range(n)]
```

```python
def zeros1(n):
    return [0 for _ in range(n)]
```

```python
def multiply_matrix(list1, list2):
    if list1 == []:
       return 0
    if type(list1[0]) == list:
        rows1 = len(list1)
        cols1 = len(list1[0])
        if type(list2[0]) == list:
              rows2 = len(list2)
              cols2 = len(list2[0])
              if cols1 != rows2:
                return 100
              else:
                result = zeros2(rows1,cols2)
                for i in range(rows1):
                    for j in range(cols2):
                        for k in range(cols1):
                            result[i][j] += list1[i][k] * list2[k][j]
        else:
             if cols1 != len(list2):
                if cols1 == 1:
                  result = zeros2(rows1,len(list2))
                  for i in range(len(list2)):
                      for j in range(rows1):
                          result[j][i] = list1[j][0] * list2[i]
                else:
                  return 200
             else:
                result = []
                for i in range(rows1):
                    sum = 0
                    for k in range(cols1):
                        sum += list1[i][k] * list2[k]
                    result.append(sum)
    else:
        if len(list1) != len(list2):
            return 300
        else:
            result = 0
            for i in range(len(list1)):
                result += list1[i] * list2[i]
    return result
```

```python
def sliceR(i,A,k):
    return [A[i][j] for j in range(k)]
```

```python
def sliceL(k,A,j):
    return [A[i][j] for i in range(k)]
```

```python
def addition_for_list(list1, list2):   # исключительно под нужды данной лабы, ни шагу в сторону
    rows1 = len(list1)                 #  так, конечно, переопределить сложение хорошо и полиморфизмы использовать в классе новом..
    rows2 = len(list2)
    if type(list1[0]) == list:
        cols1 = len(list1[0])
        cols2 = len(list2[0])
        if cols1 != cols2 or rows1 != rows2:
          return 0
        else:
          add_of_list = zeros2(rows1,cols1)
          for i in range(rows1):
              for j in range(cols1):
                add_of_list[i][j] = list1[i][j] + list2[i][j]

    else:    # подразумевается, что плохого не вводят, т.е текст или списки разных размерностей
        return [list1[i] + list2[i] for i in range(len(list1))]
    return add_of_list
```

```python
def subtraction_for_list(list1, list2):
    return addition_for_list(list1, multiply_for_list(-1, list2))
```

```python
def multiply_for_list(b,listt):
      if type(listt[0]) == list:     # у нас двумерное п-во
        rows = len(listt)
        cols = len(listt[0])
        mult_of_list = zeros2(rows,cols)
        for i in range(rows):
           for j in range(cols):
              mult_of_list[i][j] = listt[i][j]*b
      else:
        return [b*listt[i] for i in range(len(listt))]
      return mult_of_list
```

```python
def eye(n):
    A = zeros2(n,n)
    for i in range(n):
        A[i][i] = 1
    return A
```

```python
def transpose(A):
    rows = len(A)
    if type(A[0]) == list:
      cols = len(A[0])
      u = []
      for j in range(cols):
          a = []
          for i in range(rows):
            a.append(A[i][j])
          u.append(a)
    else:
        u = []
        for i in range(rows):
          u.append([A[i]])
    return u
```

```python
def decompose_LU(A):      # здесь без np.zeros ( крч перезаполнения существующей матрицы из нулей) не получится так как идёт заполнение по строке k и столбцу k
    n = len(A)
    LU = zeros2(n,n)
    for k in range(n):
        for j in range(k, n):
            LU[k][j] = A[k][j] - multiply_matrix(sliceR(k,LU,k),sliceL(k,LU,j))        #LU[k, :k] @ LU[:k, j]                # U
        for i in range(k + 1, n):
            LU[i][k] = (A[i][k] - multiply_matrix(sliceR(i,LU,k),sliceL(k,LU,k))) / LU[k][k]     #LU[i, : k] @ LU[: k, k]) / LU[k][k]  # L
    return LU
```

```python

def solution_LU(A, b):
    LU = decompose_LU(A)
   # прямой ход     Lz=b
    n = len(LU)
    z = [b[0]]

    for i in range(1,n):
        sum = 0
        for j in range(i):
            sum += LU[i][j] * z[j]
        z.append(b[i] - sum)

    # обратный ход     Ux=z
    x_inverse = [z[n-1]/ LU[i][i]]

    for i in range(n-1-1, -1, -1):
        sum = 0
        j_inverse = n - i - 1
        for j in range(i+1, n):
            sum += LU[i][j] * x_inverse[j_inverse-1]
            j_inverse= j_inverse-1
        x_inverse.append((z[i] - sum) / LU[i][i])

    x = []
    for i in range(n-1,-1,-1):
        x.append(x_inverse[i])

    return x

def determinant(A):
    LU = decompose_LU(A)
    det = 1.0
    for i in range(len(LU)):    # detA =  detL * detU
        det *= LU[i][i]
    return det

def inverse_matrix(A):
    n = len(A)
    A_inv = zeros2(n, n)
    for i in range(n):
        bb = zeros1(n)
        bb[i] = 1.0
        xx = solution_LU(A, bb)       # Ax=(0,0 ...1{i}, ...0)
        for j in range(n):
           A_inv[j][i] = xx[j]
    return A_inv
```

```python
a = 3

def f_1(x):                                         # первое уравнение системы
    return x[0]**2 + x[1]**2 - a**2

def f_2(x):                                         # второе уравнение системы
    return x[0] - np.exp(x[1]) + a


def f_1_for_plot(x):                                # смотрим на пересечение графиков, дабы определить начальную для метода точку
    return (a**2 - x**2)**0.5

def f_2_for_plot(x):
    return np.log(x + a)
```

```python
def J(x):                             # матрица Якоби
    return [
        [2*x[0],    2*x[1]],
        [1,         -np.exp(x[1])]
    ]
```

```python
def accuaracy(x, x_prev, eps):
    n = len(x)
    sum = 0
    for i in range(n):
        sum += (x[i]-x_prev[i])**2
    if np.sqrt(sum) > eps:
        return 1
    return 2
```

Достаточные условия сходимости
1. Функции $f_i(x)$ дифференцируемы
2. Матрица $J(x^{(k)})$ невырождена
3. $x_0$ лежит в достаточно маленькой $\epsilon$-окрестности искомого решения

```python
def newton_system_method(epsilon):
    x_prev = x0

    iteration = 0
    while True:

        # J(x_prev)*delta=b
        b = [-f_1(x_prev), -f_2(x_prev)]
        delta = solution_LU(J(x_prev), b)

        x = addition_for_list(x_prev , delta)            # x_prev + delta шаг метода  оказалось просто
        #print(x)
        iteration += 1
        if accuaracy(x, x_prev, epsilon) == 2:
            break
        x_prev = x
    return x, iteration
```

```python
def accuaracy(x, x_prev, eps):
    n = len(x)
    sum = 0
    for i in range(n):
        sum += (x[i]-x_prev[i])**2
    if np.sqrt(sum) > eps:
        return 1
    return 2
```

Достаточные условия сходимости метода Ньютона
1. Функции $f_i(x)$ дифференцируемы
2. Матрица $J(x^{(k)})$ невырождена
3. $x_0$ лежит в достаточно маленькой $\epsilon$-окрестности искомого решения

```python
def newton_system_method(yk, zk, f1, f2, epsilon):
    x_prev = XX0

    iteration = 0
    while True:

        # J(x_prev)*delta=b
        b = [-f1(x_prev, yk, zk), -f2(x_prev, yk, zk)]
        delta2 = solution_LU(J(x_prev), b)

        x = addition_for_list(x_prev , delta2)            # x_prev + delta шаг метода  оказалось просто
        #print(x)
        iteration += 1
        if accuaracy(x, x_prev, epsilon) == 2:
            break
        x_prev = x
    return x, iteration
```

```python
def Euler_method_implic(f, y0, dy0, x0, xn, h):        # y' = z
    x = [x0]                                      # z' = f(x,y,z)
    n = int((xn-x0)/h)
    for i in range(n):
        x.append(x[i] + h)
    y = [y0]
    z = [dy0]
    for i in range(n):
        m = newton_system_method(y[i], z[i], f_1, f_2, epsilon)   # не сломается ли от присваивания массива несколкько раз , может обнулить потом
        z.append(round(m[0][1],3))
        y.append(round(m[0][0],3))
    return x, y, z
```

```python
def func_rungekutt(x, xn, yn, zn):
     return [  - x[0] + zn + 1/4*h*x[2] + (1/4-3**0.5/6)*h*x[3],
               - x[1] + zn + 1/4*h*x[3] + (1/4+3**0.5/6)*h*x[2],
               - x[2] + f_rig(xn+(1/2-3**0.5/6)*h, yn+ 1/4*h*x[0]+(1/4-3**0.5/6)*h*x[1], zn + 1/4*h*x[2] + (1/4-3**0.5/6)*h*x[3]),
               - x[3] + f_rig(xn+(1/2+3**0.5/6)*h, yn + 1/4*h*x[1]+(1/4+3**0.5/6)*h*x[0], zn + 1/4*h*x[3] + (1/4+3**0.5/6)*h*x[2])]
```

```python
def Runge_Kutta_method_implicit(y0, dy0, x0, xn, h):
    xx = [x0]
    n = int(round((xn-x0)/h))
    for i in range(n):
        xx.append(xx[i] + h)
    yy = [y0]
    zz = [dy0]
    for i in range(n):
        def a(x):
          return func_rungekutt(x, xx[i], yy[i], zz[i])
        m = fsolve(a, [1, 1, 1, 1])        # решение нелинейного уравнения
        zz.append(0.5*(m[2]+m[3])*h+zz[i])
        yy.append(0.5*(m[0]+m[1])*h+yy[i])
    return xx, yy, zz
```
